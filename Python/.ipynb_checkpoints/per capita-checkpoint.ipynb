```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
cdeathspd = pd.read_csv('../Data/Final_Datasets/country_death_case.csv') 
cvacpd = pd.read_csv('../Data/Final_Datasets/country_vaccinations.csv') 
```

```python
cvacpd
```

```python
pop = pd.read_csv('../Data/Final_Datasets/populationedited.csv') 
```

```python
cdeathspdcapita  = pd.merge( cdeathspd, pop[['iso_code','Population']], on='iso_code')
cvacpdcapita = pd.merge( cvacpd, pop[['iso_code','Population']], on='iso_code')
```

```python
cvacpdcapita['total_vaccinations_p'] = cvacpdcapita['total_vaccinations']/cvacpdcapita['Population']
cvacpdcapita['people_vaccinated_p'] = cvacpdcapita['people_vaccinated']/cvacpdcapita['Population']
cvacpdcapita['daily_vaccinations_p'] = cvacpdcapita['people_vaccinated']/cvacpdcapita['Population']
```

```python
cvacpdcapita
```

```python
cdeathspdcapita['New_cases_p'] = cdeathspdcapita['Cumulative_cases']/cdeathspdcapita['Population']
cdeathspdcapita['New_deaths_p'] = cdeathspdcapita['Cumulative_deaths']/cdeathspdcapita['Population']
cdeathspdcapita['Cumulative_cases_p'] = cdeathspdcapita['Cumulative_cases']/cdeathspdcapita['Population']
```

```python
cdeathspdcapita
```

```python
cvacpdcapita.to_csv('country_vaccinations_p.csv')
cdeathspdcapita.to_csv('country_death_case_p.csv')
```
