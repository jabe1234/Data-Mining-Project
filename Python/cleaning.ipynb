**Cleaning**

```python
import pandas as pd
import math
```

```python
def cleanVaccinationData(file_name, new_file_name):
    # open the file
    df1 = pd.read_csv(file_name)
    
    # read headers / write to new file
    total_vaccinations = 0
    people_vaccinated = 0
    people_fully_vaccinated = 0
    daily_vaccinations_raw = 0
    daily_vaccinations = 0
    total_vaccinations_per_hundred = 0
    people_vaccinated_per_hundred = 0
    people_fully_vaccinated_per_hundred = 0
    daily_vaccinations_per_million = 0
    country_name = ''
    for index, row in df1.iterrows():
        if country_name != row["country"]:
            country_name = row["country"]
            total_vaccinations = 0
            people_vaccinated = 0
            people_fully_vaccinated = 0
            total_vaccinations_per_hundred = 0
            people_vaccinated_per_hundred = 0
            people_fully_vaccinated_per_hundred = 0
        
        if not math.isnan(row["total_vaccinations"]):
            total_vaccinations = row["total_vaccinations"]
        elif not math.isnan(row["daily_vaccinations"]):
            total_vaccinations = total_vaccinations + row["daily_vaccinations"]
            
        if not math.isnan(row["people_vaccinated"]):
            people_vaccinated = row["people_vaccinated"]
            
        if not math.isnan(row["people_fully_vaccinated"]):
            people_fully_vaccinated = row["people_fully_vaccinated"]
        
        if not math.isnan(row["total_vaccinations_per_hundred"]):
            total_vaccinations_per_hundred = row["total_vaccinations_per_hundred"]
            
        if not math.isnan(row["people_vaccinated_per_hundred"]):
            people_vaccinated_per_hundred = row["people_vaccinated_per_hundred"]
            
        if not math.isnan(row["people_fully_vaccinated_per_hundred"]):
            people_fully_vaccinated_per_hundred = row["people_fully_vaccinated_per_hundred"]
            
        
        df1.loc[df1.index[index], 'total_vaccinations'] = total_vaccinations
        df1.loc[df1.index[index], 'people_vaccinated'] = people_vaccinated
        df1.loc[df1.index[index], 'people_fully_vaccinated'] = people_fully_vaccinated
        if math.isnan(row["daily_vaccinations_raw"]):
            df1.loc[df1.index[index], 'daily_vaccinations_raw'] = daily_vaccinations_raw
        if math.isnan(row["daily_vaccinations"]):
            df1.loc[df1.index[index], 'daily_vaccinations'] = daily_vaccinations
        df1.loc[df1.index[index], 'total_vaccinations_per_hundred'] = total_vaccinations_per_hundred
        df1.loc[df1.index[index], 'people_vaccinated_per_hundred'] = people_vaccinated_per_hundred
        df1.loc[df1.index[index], 'people_fully_vaccinated_per_hundred'] = people_fully_vaccinated_per_hundred
        if math.isnan(row["daily_vaccinations_per_million"]):
            df1.loc[df1.index[index], 'daily_vaccinations_per_million'] = daily_vaccinations_per_million
        
    # convert to csv
    df1.to_csv(new_file_name)
```

```python
def cleanDeathData(file_name, new_file_name):
    # open the file
    df1 = pd.read_csv(file_name)
    
    # read headers / write to new file
    New_cases = 0
    New_deaths = 0
    country_name = ''
    for index, row in df1.iterrows():
        if country_name != row["Country"]:
            country_name = row["Country"]
        if math.isnan(row["New_cases"]):
            df1.loc[df1.index[index], 'New_cases'] = New_cases
        if math.isnan(row["New_deaths"]):
            df1.loc[df1.index[index], 'New_deaths'] = New_deaths
        
    # convert to csv
    df1.to_csv(new_file_name)
```
