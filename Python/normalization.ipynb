```python
# to convert country format
def iso3(df, column,conv):
    return df.replace({column:conv})
    
```

```python
df0 = pd.read_csv('data/vaccination-data.csv') #first vaccine dates
```

```python
# columns = ['ISO3','FIRST_VACCINE_DATE']
vacdate = df0[['ISO3','FIRST_VACCINE_DATE']]
vacdate
vacdate = vacdate.rename(columns = {'ISO3':'iso_code'}) # rename to merge with other df
vacdate
vacdate['iso_code'].nunique()
```

```python
#hw2 data country_vaccinations.csv

df1 = pd.read_csv('data/country_vaccinations.csv')
df1['iso_code'].nunique()
merged_df = pd.merge(vacdate, df1)
merged_df
# convert strings to datetime
merged_df['FIRST_VACCINE_DATE'] = pd.to_datetime(merged_df['FIRST_VACCINE_DATE'])
merged_df['date'] = pd.to_datetime(merged_df['date'])
# #calculate days since first vac
merged_df["days_since_first_vac"] = (merged_df['date'] - merged_df['FIRST_VACCINE_DATE']).dt.days


#to adjust column positions
list(merged_df.columns)

merged_df = merged_df[['iso_code',
 'FIRST_VACCINE_DATE',
'days_since_first_vac',
 'country',
 'date',
 'total_vaccinations',
 'people_vaccinated',
 'people_fully_vaccinated',
 'daily_vaccinations_raw',
 'daily_vaccinations',
 'total_vaccinations_per_hundred',
 'people_vaccinated_per_hundred',
 'people_fully_vaccinated_per_hundred',
 'daily_vaccinations_per_million',
 'vaccines',
 'source_name',
 'source_website']]


merged_df.to_csv('country_vaccinations_normalized.csv')

merged_df
```

```python
#WHO-COVID-19-global-daily-data.csv

df2 = pd.read_csv('data/WHO-COVID-19-global-daily-data.csv')
df2 = df2.rename(columns = {'Country_code':'iso_code'})
    

df2 = iso3(df2, 'iso_code',convert)

merged_df2 = pd.merge(vacdate, df2)
merged_df2
# convert strings to datetime
merged_df2['FIRST_VACCINE_DATE'] = pd.to_datetime(merged_df2['FIRST_VACCINE_DATE'])
merged_df2['Date_reported'] = pd.to_datetime(merged_df2['Date_reported'])
# calculate days since first vac
merged_df2["days_since_first_vac"] = (merged_df2['Date_reported'] - merged_df2['FIRST_VACCINE_DATE']).dt.days
merged_df2

#to adjust column positions
list(merged_df2.columns)

merged_df2 = merged_df2[['iso_code',
 'FIRST_VACCINE_DATE',
 'Date_reported',
 'days_since_first_vac',
 'Country',
 'WHO_region',
 'New_cases',
 'Cumulative_cases',
 'New_deaths',
 'Cumulative_deaths']]


merged_df2.to_csv('WHO-COVID-19-global-daily-data_normalized.csv')

merged_df2
```
